#### 구매관련 항목

```python
def getQnAPurchase():    
    driver = webdriver.Chrome()
    url = "http://help.11st.co.kr/11st/faq/FaqIndex.jsp"
    driver.get(url)
    
    quests_list = []
    for pur in purchase_list:
        next_step = '//*[@id="' + pur + '"]'
        next_step_btn = driver.find_element_by_xpath(next_step)
        next_step_btn.click()

        soup = BeautifulSoup(driver.page_source, 'lxml')

        total_quests = len(soup.find('div', {'class': 'paging'}).find("span").findAll("a")) # 총페이지 갯수
        lenth = len(soup.find('tbody').findAll('td', {'class': 'tdLeft'})) # 한페이지의 qna 갯수 : 15개

        quests = soup.find('tbody').findAll('td', {'class': 'tdLeft'})

        try:
            for page in range(1, total_quests+1):
                next_page = '//*[@id="contentContainer"]/div[2]/span/a[' + str(page) + ']'
                next_page_btn = driver.find_element_by_xpath(next_page)
                next_page_btn.click()

                soup = BeautifulSoup(driver.page_source, 'lxml')
                quests = soup.find('tbody').findAll('td', {'class': 'tdLeft'})

                for i in range(lenth):
                    quest = quests[i].text
                    quests_list.append(quest)
            print(len(quests_list))
        except:
            print(len(quests_list))
    
    qna_df = pd.DataFrame()
    qna_df["내용"] = quests_list
    
    return qna_df    
```
